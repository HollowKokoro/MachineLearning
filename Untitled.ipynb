```python
import plotly.express as px
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_csv('./Most Popular Programming Languages from 2004 to 2020 V3.csv')
dateFormatted = pd.to_datetime(df.Date, format='%B %Y', errors='raise')
df.Date = dateFormatted
```

```python
fig = px.line(df, x='Date', y=list(df.columns))
fig.show()
```

```python
res = []
for x in df.columns:
    data = df[x].to_numpy(dtype=float)
    res.append(data)
```

```python
#AutoReg
from statsmodels.tsa.ar_model import AutoReg
autoregAll = []
for x in res:
    data = AutoReg(endog = x, lags = [1, 11, 12]).fit()
    autoreg = data.predict(len(res), len(res))
    autoregAll.append(autoreg)
print(autoregAll)
```

```python
#MA
from statsmodels.tsa.arima.model import ARIMA
maAll = []
for x in res:
    data = ARIMA(endog = x, order=(0, 0, 1)).fit()
    ma = data.predict(len(res), len(res))
    maAll.append(ma)
print(maAll)
```

```python
#ARMA
from statsmodels.tsa.arima.model import ARIMA
armaAll = []
for x in res:
    data = ARIMA(endog = x, order=(2, 0, 1)).fit()
    arma = data.predict(len(res), len(res))
    armaAll.append(arma)
print(armaAll)
```

```python
#ARIMA
from statsmodels.tsa.arima.model import ARIMA
arimaAll = []
for x in res:
    data = ARIMA(endog = x, order=(1, 1, 1)).fit()
    arima = data.predict(len(res), len(res))
    arimaAll.append(arima)
print(arimaAll)
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
sesAll = []
for x in res:
    data = SimpleExpSmoothing(x).fit()
    ses = data.predict(len(res), len(res))
    sesAll.append(ses)
print(sesAll)
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
esAll = []
for x in res:
    data = ExponentialSmoothing(x).fit()
    es = data.predict(len(res), len(res))
    esAll.append(es)
print(esAll)
```

```python
autoregAll = pd.DataFrame(autoregAll)
maAll = pd.DataFrame(maAll)
autoregAll.join(maAll, how='right')
autoregAll
```
